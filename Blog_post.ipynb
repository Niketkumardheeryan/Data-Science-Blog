## Importing the Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 
data=pd.read_csv('Startups.csv')        #https://www.kaggle.com/johndasilva/diabetes

data.head()                           #highlight of data-set
```

### information about dataset used

```python
# gettinng information about data-set


print(data.info())          # no null values found

print(data.shape)           #no. of Rows and columns   
```

```python
data  =  data.drop(['State'], axis=1)           #removing unessential feature 

data.describe()
```

```python
data.isnull().sum()              # getting the sum of total null values
```

```python
data.head()

```

```python
data.plot.hist()                  # visualized data
```

# Multiple linear regression model

```python
# dividing the data set into dependent or independent variabes

x=data.iloc[:, 0:3].values           #Dependent variable

y=data.iloc[:, 3].values             #independent variable 


from sklearn.model_selection  import train_test_split 

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=0)  #splitting of ddata set


# creating Multiple linear regression Model

from sklearn.linear_model import LinearRegression



reg_obj = LinearRegression()

reg_obj.fit(xtrain,ytrain)

y_pred = reg_obj.predict(xtest)



# visualizing the actual result vs predicted result


ax1 = sns.distplot(ytest, hist=False, color="r", label="Actual Value")

sns.distplot(y_pred, hist=False, color="b", label="Fitted Values" , ax=ax1)
```

```python






sns.pairplot(data);          # more information about data set




```

# Score

```python
from sklearn.metrics import r2_score

score=r2_score(ytest,y_pred)
score
```
